###  Spotify Song Recommendation (Spotify makes new song recommendations to users every week based on the songs that are currently saved in their playlists )

```python
import random,string
songids=['SONGID'+str(i).zfill(2) for i in range(1,21)]
userid=[''.join('USER')+str(random.randint(10,99)).zfill(2) for i in range(100)]
playlist={i:j for i,j in zip(userid,[random.sample(songids,random.randint(4,7)) for i in range(len(userid))])}
playlist
```

```python
## for eg: new_user_song is the song playlist saved by the new user

new_user_songs = ['SONGID08','SONGID09','SONGID12']

# Finding playlists that contain all the new user's songs
matching_playlists = []

for user, songs in playlist.items():
    if set(new_user_songs).issubset(set(songs)):
        matching_playlists.append(user)

# Checking if there are matching playlists found
if len(matching_playlists) > 0:
    selected_playlist = random.choice(matching_playlists)  # Randomly select a playlist from the matching playlists
    remaining_songs = list(set(playlist[selected_playlist]) - set(new_user_songs)) # Get the remaining songs from the selected playlist
    print("Recommended Songs for the new user:")
    for song in remaining_songs:
        print(song)
else:
    print("No matching playlists found. Unable to provide song recommendations.")
```
